### theory of large numbers

when we repeat the experiment for more number of times , we may get expected probability

```python
# lets repeat the experiment 1 milllion times

one_m = np.random.randint(1, 7, 1000000)  # dice: 1 - 6; 7 - execluded
sns.histplot( x = one_m, bins = 6)
```

### now lets see if we can make the abvoe a normal dsitrubtion

```python
np.mean(one_m)
```

```python
np.std(one_m)
```

# lets repeat the simulation on large data

```python
one_m = np.random.randint(1, 7, 1000000)
mean_1M = one_m.reshape(250000, 4).mean(axis = 1)
sns.histplot( x = mean_1M, bins = 6 )
```

### as per central limit theorem, check if the resultant is mean is equal to the mean of the population

```python
np.mean(one_m)
```

```python
np.mean(mean_1M)
```

#### the above confirms the correctness of the central limit theorem

## Now, lets verify the std. dev

```python
np.std(mean_1M)
```

```python
# std.dev/sqrt(n)   n=size of sample

np.std(one_m)/np.sqrt(4)
```

#### the above confirms the correctness of the central limit theorem

i.e. std.dev of population = std.dev of sample means

# bi-model distribution  -- result of overalpping one normal distributino with other one

it has 2 modes

```python
# create a normal distribution which has mean-100, std dev=5 and create 1000 such samples i.e. fetch 1000 samples which have
# mean 100 and std dev as 5

a = np.random.normal( 100, 5, 1000 )   # mean=100, std dev=5 and create 1000 such samples
b = np.random.normal( 80, 5, 1000 )    # mean=80, std dev=5 and create 1000 such samples
```

```python
sns.distplot(a)
```

```python
sns.distplot(b)
```

## now append both samples 1000 + 1000 = 2000 to generate a bimodel distribution

```python
bm_data = np.append(a, b)
```

```python
sns.distplot(bm_data)
```

# now shuffile the distribution to avoid any sort of biasness

```python
np.random.shuffle(bm_data)
```

### now, lets apply central limit theorem

# now lets take random samples of 2000 records and plot their mean

```python
# sample size = 2

bm2 = bm_data.reshape(1000, 2).mean( axis = 1)   
sns.distplot(bm2)
```

```python
# sample size = 4

bm4 = bm_data.reshape(500, 4).mean( axis = 1)   
sns.distplot(bm4)
```

```python
# sample size = 10   --- almost a normal distribution

bm10 = bm_data.reshape(200, 10).mean( axis = 1)   
sns.distplot(bm10)
```

### now lets validate central limit theorm by comparing mean and std deviation of population and central limit distribution

#### check mean

```python
np.mean(bm_data)
```

```python
np.mean(bm10)
```

#### check std dev

```python
np.std(bm_data)/np.sqrt(10)
```

```python
np.std(bm10)
```
